# Notebook 01: Baseline Performance Analysis
This notebook analyzes the baseline performance of **Nemotron-3 Nano 30B** in FP8 and BF16 modes on the H100 GPU (Node g18).

```python
import os
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set_theme(style="whitegrid")
```

### Load Consolidated Benchmark Results

```python
csv_path = "../results/consolidated_benchmark_results.csv"
if os.path.exists(csv_path):
    df = pd.read_csv(csv_path)
    print(f"Loaded {len(df)} records.")
else:
    print("Results CSV not found. Run analysis/parse_results.py first!")
    # Generate dummy data for illustration
    data = []
    for con in [1, 4, 8, 16]:
        data.append({"Concurrency": con, "Expert Weight": 0.0, "throughput_tokens_sec": 145.0 * (1.0 + 0.4 * con**0.5), "ttft_p50_ms": 45.0 + con * 2.0, "itl_p50_ms": 8.5 + con * 0.2, "kv_cache_hit_rate": 0.82})
    df = pd.DataFrame(data)
```

### Plot Baseline Throughput vs Concurrency

```python
baseline_df = df[df["Expert Weight"] == 0.0]

plt.figure(figsize=(8, 5))
plt.plot(baseline_df["Concurrency"], baseline_df["throughput_tokens_sec"], marker='o', linewidth=2, color='royalblue')
plt.title("Baseline Throughput (expert_weight=0.0)")
plt.xlabel("Concurrency")
plt.ylabel("Throughput (tokens/sec)")
plt.xticks(baseline_df["Concurrency"].unique())
plt.tight_layout()
plt.show()
```
