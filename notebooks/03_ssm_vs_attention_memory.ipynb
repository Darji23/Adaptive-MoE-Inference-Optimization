# Notebook 03: Mamba-2 SSM vs Self-Attention Memory Traffic
This notebook evaluates the differences in memory bandwidth load of Mamba-2 SSM recurrent state updates vs Standard Attention KV fetches.

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set_theme(style="whitegrid")
```

### Load Memory Bandwidth Analysis Report

```python
report_path = "../results/profiling/ssm_bandwidth_analysis.csv"
if os.path.exists(report_path):
    df = pd.read_csv(report_path)
else:
    # Simulated stats
    df = pd.DataFrame([
        {"Layer Type": "Attention (with KV Cache)", "Latency (ms)": 0.125, "HBM Memory Transferred (MB)": 256.0, "Effective Bandwidth (GB/s)": 2048.0, "Bottleneck Category": "Memory-Bandwidth Bound (HBM Bound)"},
        {"Layer Type": "Mamba-2 SSM (Recurrent State)", "Latency (ms)": 0.015, "HBM Memory Transferred (MB)": 4.2, "Effective Bandwidth (GB/s)": 280.0, "Bottleneck Category": "Compute/Latency Bound (Sequential Dependency)"}
    ])
df
```

### Plot Bandwidth Comparison

```python
plt.figure(figsize=(8, 4))
sns.barplot(data=df, x="Layer Type", y="Effective Bandwidth (GB/s)", palette="Set2")
plt.title("Effective HBM Memory Bandwidth (H100)", fontsize=13, fontweight='bold')
plt.ylabel("Bandwidth (GB/s)")
plt.tight_layout()
plt.show()
```
